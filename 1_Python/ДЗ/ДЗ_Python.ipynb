### Задание 1


Из списка выбрать индексы тех элементов, которые в сумме дают результат. Использовать одни и теже элементы дважды нельзя. Порядок вывода значения не имеет.

**_Пример 1_**

Входящий список: nums = [2,7,11,15], результат - 9

Выход: [0,1]

Объяснение: поскольку 2 + 7 = 9, а позиция элементов 0 и 1

Входящий список: nums = [3,2,4], результат = 6

Выход: [1,2]

### Задание 2


В данной строке, включающей слова и пробелы, подсчитать количество символов в последнем слове. Слово - максимальная подстрока, внутри которой нет пробелов

**_Пример 1:_**

Входящая строка: s = "Hello World"

Выход: 5

Объяснение: Последнее слово "World" и его длина равна 5.

**_Пример 2:_**

Входящая строка: s = " fly me to the moon "

Выход: 4

Объяснение: Последнее слово "moon" и его длина равна 4.


```python
inp_str = "   fly me  , to; :  the moonlight !!! "
# Заменяем символы, которые не будем учитывать в словах, на пробелы
whitespaced_str = inp_str.translate(inp_str.maketrans(",.:;!?", "      "))
# Разбиваем по пробелам на слова, берём длину первого с конца
len(whitespaced_str.split()[-1])
```

### Задание 3


Дано целое число INT. Напишите функцию, которая возвращает TRUE, если число ПАЛИНДРОМ(одинаково как сначала в конец, так и наоборот), и FALSE в противном случае.

**_Пример 1:_**

Вход: 121

Выход: TRUE

**_Пример 2:_**

Вход: 345

Выход: FALSE


```python
# Отошёл от ТЗ по формату вывода, т.к. так удобнее проверять
test_nums = [121, 345, 1234321, 1234627]

for num in test_nums:
    print(num, end='\t')
    # Преобразуем число в строку, т.к. задача проще решается сравнением символьных строк, чем мат. преобразованиями
    num_str = str(num)
    # Определяем количество символов для сравнения
    half_len = len(num_str) // 2
    # Сравниваем первую половину и развёрнутую вторую половину строки
    if (num_str[:half_len] == num_str[-1:-half_len-1:-1]):
        print("TRUE")
    else:
        print("FALSE")
```
