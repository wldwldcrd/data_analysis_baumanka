### Задание 1

Из списка выбрать индексы тех элементов, которые в сумме дают результат. Использовать одни и теже элементы дважды нельзя. Порядок вывода значения не имеет.

***Пример 1***

Входящий список: nums = [2,7,11,15], результат - 9

Выход: [0,1]

Объяснение: поскольку 2 + 7 = 9, а позиция элементов 0 и 1

***Пример 2:***

Входящий список: nums = [3,2,4], результат = 6

Выход: [1,2]

Объяснение: поскольку 2 + 4 = 6, а позиция элементов 1 и 2


```python
# Ограничение: работа производится только с положительными числами (то есть ни резульат ни проверемый массив не должен содержать отрицательные значения)
# Отображение рассчитанных вариантов не убрал. При необходимости можно не отображать расситанные вараинты, а отобразить только уникальные комбинации индексов исходного списка
import itertools
nums = [2, 7, 11, 20, 15, 5]
result = 20
options = list(itertools.permutations(nums))
if sum(nums) < result:
    print('Результат введен некорректно. Результат больше суммы всех элементов списка. Сумма элементов списка:', sum(
        nums), 'введенный результат:', result)
else:
    for step in options:
        controlValue = 0
        controlListIndex = []
        controlList = []
        for stepDown in step:
            controlValue = controlValue + stepDown
            controlListIndex.append(step.index(stepDown))
            controlList.append(stepDown)
            p = []
            if controlValue == result:
                for x in controlList:
                    p.append(nums.index(x))
                print('Корректная комбинация:', step)
                print('Используемые значения:', controlList)
                print('Индексы исходного списка:', p, '\n')
                # print('Индексы используемых значений: ', controlListIndex, '\n')
                continue
```

### Задание 2

В данной строке, включающей слова и пробелы, подсчитать количество символов в последнем слове. Слово - максимальная подстрока, внутри которой нет пробелов


***Пример 1:***

Входящая строка: s = "Hello World"

Выход: 5

Объяснение: Последнее слово "World" и его длина равна 5.


***Пример 2:***

Входящая строка: s = "   fly me   to   the moon  "

Выход: 4

Объяснение: Последнее слово "moon" и его длина равна 4.

```python
#Должны ли исключаться знаки препинания? Что делаем с цифрами?

#простой вариант с просто последним элементом
s = " Hello World   "
s = s.strip().split(' ')
print(len(s[len(s)-1]))

#вариант с исключением знаков препинания (цифры не трогаем)
import string
s1 = " Hello Worlds!!!!!! !!!!!  "
s1 = s1.translate(str.maketrans('', '', string.punctuation))
s1 = s1.strip().split(' ')
print(len(s1[len(s1)-1]))
```

### Задание 3


Дано целое число INT. Напишите функцию, которая возвращает TRUE, если число ПАЛИНДРОМ(одинаково как сначала в конец, так и наоборот), и FALSE в противном случае.

***Пример 1:***

Вход: 121

Выход: TRUE

***Пример 2:***

Вход: 345

Выход: FALSE



```python
def turn(standard):
  value1 = standard
  invertedValue = 0
  while standard > 0:
    digit = standard % 10
    standard = standard // 10
    invertedValue = invertedValue * 10
    invertedValue = invertedValue + digit

  if value1 == invertedValue:
    return True
  else:
    return False


value = int(input("Проверяемое число: "))
turn(value)
```
