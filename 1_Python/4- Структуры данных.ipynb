### Словари - тип данных, представляющий собой коллекцию пар «ключ-значение». В рамках одного словаря ключи не повторяются.

```python
dict_1 = dict({'John': 29, 'Simon': 45, 'Jeniffer': 33,
              'Morgan': 55, 'Simon': 55})
dict_1
```

```python
dict_1 = {'John': 29, 'Simon': 45, 'Jeniffer': 33, 'Morgan': 55, 'Simon': 55}
```

```python
dict_1['Ann'] = 45
dict_1
```

```python
dict_1['Morgan']
```

```python
dict_1['Ann'] = 55
```

```python
dict_1
```

```python
dict_1.items()
```

```python
dict_1.keys()
```

```python
dict_1.values()
```

```python
dict_1.pop('John')
dict_1
```

```python
lst_1 = [1, 2, 1, 4, 5, 5, 3, 2]
tuple_1 = tuple(lst_1)
tuple_1
```

```python
tuple_2 = (1,)
tuple_2
```

```python
tuple_1.count(1)
```

```python
var = tuple_1[0]
var
```

```python
tuple_1.pop()
```

```python
type(tuple_1)
```
