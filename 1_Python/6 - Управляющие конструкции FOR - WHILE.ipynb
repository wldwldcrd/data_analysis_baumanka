### Цикл WHILE


```python
iter = 100
step = 5
i = 1

while i <= iter:
    print(f"{i}")
    i += step
```

```python
step = -5
i = -100

while i <= 0:
    print(f"{i}")
    i = i - step
```

```python
str_1 = 'Python'
i = 0
while (i <= len(str_1) - 1):
    # i +=1
    # if (i == 8):
    #   break
    # elif(i == 5):
    #   continue
    # else:
    print(str_1[i])
    i += 1
```

### FOR

```python
list_a = [1, 2, 3, 4, 5, 6]
list_b = ('p', 'y', 't', 'h', 'o', 'n')


for x in list_b:
    print(x)
```

```python
list_b = "Python"
for x in list_b:
    print(x)
```

```python
for i in range(0, 6):  # 0 .. 4
    print(list_a[i])
```

```python
for i in range(1, 10):  # 1 .. 9
    print(list_b[i])
```

#### List comprehension

```python
lst_1 = [1, 2, 3, 4, 5]
for x in lst_1:
    print(x ** 2)
```

```python
[x ** 2 for x in lst_1]
```

### Тернарный оператор
#### Значение_1 IF условие ELSE Значение_2

```python
int_1 = 26
int_2 = 27

res = int_1 if int_1 < int_2 else int_2
print(res)
```

```python
a = "int_1 " + ("even " if int_1 % 2 == 0 else "odd ") + "number"
print(a)
```
