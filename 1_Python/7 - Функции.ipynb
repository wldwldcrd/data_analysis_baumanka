### Декораторы

```python
# First approach: non-parametric
def func_decorator(func):
    def wrapper():
        print("Before FUNC")
        func()
        print("After FUNC")
    return wrapper
```

```python
@func_decorator
def decor():
    print("I am decorator FUNC")
```

```python
decor()
```

```python
# Second: parametric
def func_decorator(func):
    def wrapper(*args, **kwargs):
        print("Before FUNC")
        func(*args, **kwargs)
        print("After FUNC")
    return wrapper
```

```python
@func_decorator
def decor(title):
    print(f"I like {title}")
```

```python
decor('Linux')
```
