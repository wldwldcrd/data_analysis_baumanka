### IF - ELSE


```python
x = 5
if (x > 5):
    print('More')
else:
    print('Less')
```

```python
int_1 = 5

if (int_1 >= 5):
    print("Больше или равно")
else:
    print("Меньше")
```

```python
int_1 = 4
if (not (type(int_1) is int)):
    print("Int")
```

```python
str_1 = "Time is money - that is a saying"
if (len(str_1) > 20):
    print("It's a huge line")
else:
    print("Tiny one")
```

```python
a = 6
b = 5

if a > b:
    a, b = b, a

print(a, b)
```

```python
int_1 = 5
int_2 = 22

x = float(input("x: "))

if int_1 <= x <= int_2:
    print("Within")
else:
    print("You are not in our circle")
```

```python
list_ = [1, 2, 3, 4, 5]

x = float(input("x: "))

if x in list_:
    print("Welcome on board")
```

### Вложенные IF


```python
int_1 = 3
int_2 = 7
int_3 = 5

if int_1 > int_2:
    if int_2 > int_3:
        print("INT_1 is MAX")
    else:
        print("INT_3 is MAX")
else:
    if int_2 > int_3:
        print("INT_2 is MAX")
    else:
        print("INT_3 is MAX")
```

```python
max(int_1, int_2, int_3)
```

```python
a = int(input("Number from 1 to 4: "))

if a == 1:
    print("Your lucky language is Python")
elif a == 2:
    print("Your lucky language is Java")
elif a == 3:
    print("Your lucky language is C")
elif a == 4:
    print("Your lucky language is JavaScript")
else:
    print("Not a chance")
```
